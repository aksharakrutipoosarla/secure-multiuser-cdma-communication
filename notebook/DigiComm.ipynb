```python
import matplotlib.pyplot as plt

def dif_man(bitstring, initial_level=1, start_transition_means=0):
    if not isinstance(bitstring, str) or any(b not in "01" for b in bitstring):
        raise ValueError("Input must be a string of '0' and '1' characters (e.g. '10110010').")

    level = 1 if initial_level >= 0 else -1
    signal = []

    for bch in bitstring:
        bit = int(bch)

        if bit == start_transition_means:
            level = -level   

        signal.append(level)

        level = -level

        signal.append(level)

    return signal


def plot_dif_man(bitstring, initial_level=1, start_transition_means=0):
    signal = dif_man(bitstring, initial_level, start_transition_means)

    x = list(range(len(signal)))

    plt.figure(figsize=(10, 2.6))
    plt.step(x, signal, where='post')
    plt.ylim(-1.5, 1.5)
    plt.yticks([-1, 1])
    plt.xlabel("Time (half-bit samples; 2 samples = 1 bit)")
    plt.title("Differential Manchester Encoding  (start transition ⇒ {})".format(start_transition_means))
    plt.grid(axis='x', linestyle=':', alpha=0.5)

    nbits = len(bitstring)
    bit_positions = [i*2 + 1 for i in range(nbits)]
    plt.xticks(bit_positions, list(bitstring))

    for k in range(nbits + 1):
        plt.axvline(k*2, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)   # bit boundaries
    for k in range(nbits):
        plt.axvline(k*2 + 1, color='gray', linestyle=':', linewidth=0.5, alpha=0.35) # mid-bit transitions

    plt.xlim(-0.5, 2*nbits - 0.5)
    plt.tight_layout()
    plt.show()
plot_dif_man("10110010", initial_level=1, start_transition_means=0)
```
